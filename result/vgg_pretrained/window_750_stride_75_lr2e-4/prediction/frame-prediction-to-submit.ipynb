```python
from pathlib import Path
```

```python
train_events = Path("../input/grasp-and-lift-eeg-detection/train/").glob("*events.csv") 
```

```python
train_events_df = pd.concat([pd.read_csv(path) for path in train_events], 
                            axis=0)
```

```python
train_events_df.shape
```

```python
train_predicted_df = pd.read_csv("../input/window-prediction-to-frame-predictions/train_frame_prediction.csv") 
```

```python
train_events_df[train_events_df.id == "subj12_series12_153650"]
```

```python
train_predicted_df[-1000:]
```

```python
train_diff = set(train_events_df.id) - set(train_predicted_df.id)  
```

```python
train_events_df[:10]
```

```python
len(train_diff)
```

```python
list(train_diff)[:10]
```

```python
rows = []
for id_ in train_diff:
    row = {col: 0 for col in train_events_df.columns[-6:]} 
    row["id"] = id_
    rows.append(row)
```

```python
len(rows)
```

```python
train_predicted_df.shape
```

```python
train_predicted_df = train_predicted_df.append(rows, ignore_index=True)
```

```python
train_predicted_df.shape
```

```python
train_predicted_df[-10:]
```

```python
import re
```

```python
ID_PATTERN = r"subj(?P<subject>\d+)_series(?P<series>\d+)_(?P<frame>\d+)"
```

```python
extracted = train_predicted_df["id"].str.extract(ID_PATTERN)
```

```python
extracted[:10]
```

```python
for col in extracted.columns:
    train_predicted_df[col] = extracted[col].astype("int")
```

```python
train_predicted_df = train_predicted_df.sort_values(by=["subject", "series", "frame"], axis=0)
```

```python
train_predicted_df[:10]
```

```python
extracted = train_events_df["id"].str.extract(ID_PATTERN)
```

```python
for col in extracted.columns:
    train_events_df[col] = extracted[col].astype("int")
```

```python
del extracted
```

```python
import gc
gc.collect()
```

```python
train_events_df = train_events_df.sort_values(by=["subject", "series", "frame"], axis=0)
```

```python
from sklearn.metrics import roc_auc_score
```

```python
train_predicted_df.shape[0] 
```

```python
train_events_df.shape[0] 
```

```python
set(train_predicted_df.id) - set(train_events_df.id) 
```

```python
train_predicted_df = train_predicted_df[train_predicted_df.id != 'subj4_series2_208724']
```

```python
target_columns = train_events_df.columns.tolist()[1:7]
```

```python
target_columns
```

```python
[(id1, id2) for id1, id2 in zip(train_events_df.id, train_predicted_df.id) if id1 != id2]
```

```python
roc_auc_score(train_events_df[target_columns].values, 
              train_predicted_df[target_columns].values)
```

```python
del  train_events_df
```
